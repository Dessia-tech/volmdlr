```python
from IPython.display import HTML

import math

import volmdlr as vm
import volmdlr.core as vmc
import volmdlr.primitives3d as p3d
```

```python
frame = vm.Frame3D(origin=vm.Point3D(1, 1, 1),
                   u=vm.Vector3D(2, 0, 0),
                   v=vm.Vector3D(0, 0.5, 0),
                   w=vm.Vector3D(0, 0, 1))
block1 = p3d.Block(frame=frame, color=(1, 0, 0))
block1.save_babylonjs_to_file(filename='/tmp/block1')
HTML(filename='/tmp/block1.html')
```

```python
frame = vm.Frame3D(origin=vm.O3D,
                   u=vm.Vector3D(2, 0, 0),
                   v=vm.Vector3D(0, 0.5, 0),
                   w=vm.Z3D)
block2 = p3d.Block(frame=frame,
                   color=(0, 0, 1))
block2.rotation(center=vm.O3D,
                axis=vm.Y3D,
                angle=math.pi/4,
                copy=False)
block2.rotation(center=vm.O3D,
                axis=(vm.X3D + vm.Z3D)/2**0.5,
                angle=math.pi/2,
                copy=False)
block2.translation(offset=vm.Z3D, copy=False)
volume_model = vmc.VolumeModel([block1, block2])
volume_model.save_babylonjs_to_file(filename='/tmp/volume_model')
HTML(filename='/tmp/volume_model.html')
```

```python
unions = block1.union(block2)
print(unions)

unions[0].save_babylonjs_to_file(filename='/tmp/union')
HTML(filename='/tmp/union.html')
```

```python
intersections = block1.intersection(block2)
print(intersections)

intersections[0].save_babylonjs_to_file(filename='/tmp/intersection')
HTML(filename='/tmp/intersection.html')
```

```python
block1.alpha = 0.2
block2.alpha = 0.2
intersections[0].color = (0, 1, 0)
intersection_volume_model = vmc.VolumeModel(primitives=[block1, block2, intersections[0]])

intersection_volume_model.save_babylonjs_to_file(filename='/tmp/intersection_volume_model')
HTML(filename='/tmp/intersection_volume_model.html')
```
